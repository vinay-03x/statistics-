## 2) can you tell me the distribution of vechicles based on year of registration with the help of a plot

```python
v["yearOfRegistration"] = v["yearOfRegistration"].where((v["yearOfRegistration"] >= 1950) & (v["yearOfRegistration"] <= 2023),v["yearOfRegistration"].mode()[0])
plt.figure(figsize=(20,10))
sns.histplot(data=v,x=v["yearOfRegistration"])
plt.title("Distribution of Vechicles Based on Year of Registration")
plt.grid(linestyle='--')
plt.show()
```

## 3) Create a plot based on the Variation of the price range by the vehicle type

```python
v["vehicleType"].value_counts()
```

```python
avg_type = v.groupby("vehicleType")["price"].mean()
avg_type = avg_type.sort_values()
plt.figure(figsize=(15,8))
sns.barplot(x=avg_type.index , y=avg_type.values)
plt.ylabel("Avg price")
plt.title("Price Range Variation by Vehicle Type (Bar Plot)")
plt.show()
```

## 3)What is the marginal probability of private seller

### Calculate the total number of listings

```python
total_listings = len(v)
```

### Calculate the number of listings with private sellers

```python
private_seller_count = (v['seller'] == 'privat').sum()
```

### Calculate the marginal probability of private sellers

```python
marginal_probability_private = private_seller_count / total_listings
print(f"Marginal Probability of Private Seller: {marginal_probability_private:.2f}")
```

# Analysis3

### What is the Average price of vehicle by fuel type and gearbox type.Give a plot

```python
avg_price=v.groupby(['fuelType', 'gearbox'])['price'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=avg_price, x='fuelType', y='price', hue='gearbox')
plt.title('Average Price by Fuel Type and Gearbox')
plt.show()
```

### What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot

```python
avg=v.groupby(['vehicleType', 'gearbox'])['powerPS'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=avg, x='vehicleType', y='powerPS', hue='gearbox')

plt.title('Average Power by Vehicle Type and Gearbox')
plt.show()
```

### What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
data=v.groupby(['brand', 'vehicleType'])['price'].mean().unstack()
plt.figure(figsize=(20, 15))
sns.heatmap(data, annot=True, fmt=".2f", linewidths=.5)

plt.title('Average Price by Brand and Vehicle Type (Heatmap)')
plt.show()
```
